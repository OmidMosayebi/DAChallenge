```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("2023_technical_challenge_data_analyst (1).csv")
```

```python
df.head()
```

```python
df.fillna(0,inplace=True)
```

```python
for i in df['LAST_X_WEEKS'].unique():
    print(i,df[df['LAST_X_WEEKS']==i]['PLACES_RECOMMENDED'].describe())
```

```python
l_th = []
for i in df['LAST_X_WEEKS'].unique():
    l_th.append([i,df[(df['LAST_X_WEEKS']==i) ]['THANKYOUS_RECEIVED'].sum(),
                 df[(df['LAST_X_WEEKS']==i)& (df['THANKYOUS_RECEIVED']>=1)]['THANKYOUS_RECEIVED'].count(),
                 df[(df['LAST_X_WEEKS']==i)& (df['THANKYOUS_RECEIVED']>=1)]['THANKYOUS_RECEIVED'].count() / 200,
                 df[(df['LAST_X_WEEKS']==i)]['THANKYOUS_RECEIVED'].sum() /
                 df[(df['LAST_X_WEEKS']==i)& (df['THANKYOUS_RECEIVED']>=1)]['THANKYOUS_RECEIVED'].count()])
    

df_t = pd.DataFrame(l_th,columns=['X_WEEKS','ITEM_COUNT','USER_COUNT','ENGAGEMENT','PERCENTAGE_TOTAL'])
```

```python
l_th = []
for i in df['LAST_X_WEEKS'].unique():
    l_th.append([i,df[(df['LAST_X_WEEKS']==i) ]['EVENTS_CREATED'].sum(),
                 df[(df['LAST_X_WEEKS']==i) & (df['EVENTS_CREATED']>=1)]['EVENTS_CREATED'].count(),
                 df[(df['LAST_X_WEEKS']==i) ]['EVENTS_CREATED'].sum() /
                 df[(df['LAST_X_WEEKS']==i) & (df['EVENTS_CREATED']>=1)]['EVENTS_CREATED'].count(),
                 df[(df['LAST_X_WEEKS']==i) & (df['EVENTS_CREATED']>=1) &(df['EVENT_PARTICIPANTS']>=1)]['EVENT_PARTICIPANTS'].sum(),
                df[(df['LAST_X_WEEKS']==i) & (df['EVENTS_CREATED']>=1) &(df['EVENT_PARTICIPANTS']>=1)]['EVENT_PARTICIPANTS'].sum() / 
                 df[(df['LAST_X_WEEKS']==i) ]['EVENTS_CREATED'].sum(),
                df[(df['LAST_X_WEEKS']==i) & (df['EVENTS_CREATED']>=1)]['EVENT_PARTICIPANTS'].count() / 200
                ])
df_t = pd.DataFrame(l_th,columns=['X_WEEKS','EVENT_COUNT','USER_COUNT','EVENT_RATE','PARTICIPANT_COUNT','EVENT_PAR_RATE','ENGAGEMENT'])
```
